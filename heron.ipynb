```python
"""Part One: Heron Method"""
import os

a = 2
N = 30
x0_list = [3, 5, 7]

class Heron:
    
    #starting values
    def __init__(self):
        self.xlists = [[], [], []]
        self.errorlists = [[], [], []]
        
    def heron_func(self):
        for b in range(len(x0_list)):
            x = x0_list[b]
            #print('x is', x, 'b is', b)
            for n in range(N):
                x_plus = (0.5)*(x + a/x)
                self.xlists[b].append(x_plus)
                x = x_plus
        return(self.xlists)
    
    def relative_error(self):
        for b in range(len(x0_list)):
            xnlist = self.xlists[b]
            squared_numbers = list(map(lambda x: x**2, xnlist))
            for n in range(N):
                xn = squared_numbers[n]
                error = abs(xn - a)/a
                self.errorlists[b].append(error)  
        return(self.errorlists)

heron = Heron()
result = heron.heron_func()  # a=25 and N=10 iterations
error = heron.relative_error()  #squared error numbers
```

```python
import matplotlib.pyplot as plt
import numpy as np

pathdir = 'plots/individual_xn_plots'
if not os.path.exists(pathdir):
    os.makedirs(pathdir)

#individual plots
for i in range(len(x0_list )):
    x0 = x0_list[i]         #this is starting point for x, x0
    xn = result[i]         #this is the calculated herons method xns
    squared_numbers = list(map(lambda x: x**2, xn))
    itera = np.arange(0, N, 1)      #the iteration number

    plt.plot(squared_numbers, itera, label=r"estimation for $(\sqrt{x_{n+1}})^2$", color='blue')
    plt.plot(xn, itera, label=r"estimation for $\sqrt{x_{n+1}}$", color='green')

    #scatter plots with zorder=3 so it overlays on top of the line graph. 
    plt.scatter(squared_numbers, itera, color='black', s=18, label="_nolegend_", zorder=3)
    plt.scatter(xn, itera,   color='orange', label="_nolegend_" , s=8, zorder=3)
    plt.axvline(x=2, color='red', linestyle='--', label='x=2')
    plt.axvline(x=1.41421356237, color='yellow', linestyle='--', label=r'x=$\sqrt{2}$')

    
    plt.ylim(-0.1, 5) 
    
    plt.title(fr"Heron Method for a=2: starting value $x_0 = {x0}$")      # Title of the plot
    plt.xlabel(r"Input $x_{n+1} = {\frac{1}{2}(x_n + \frac{2}{x_n}})$")                      # Label for x-axis
    plt.ylabel("Iteration number, n")  
    plt.legend()
        
    #SAVE the plot to the directory
    save_path = os.path.join(pathdir, fr'heronplot{x0}.png')
    plt.savefig(save_path)  #save the figure as a .png file

    print(f"Plot saved to: {save_path}")

    #show the plot
    plt.show()

    #clear figure so plots dont overlap.
    plt.clf()
    
    
    plt.show()
    
    
```
